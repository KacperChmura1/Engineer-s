```python
# Lista nazw województw z małymi literami
wojewodztwa = ["dolnośląskie", "kujawsko-pomorskie", "lubelskie", "lubuskie", "łódzkie", "małopolskie",
               "mazowieckie", "opolskie", "podkarpackie", "podlaskie", "pomorskie", "śląskie",
               "świętokrzyskie", "warmińsko-mazurskie", "wielkopolskie", "zachodniopomorskie"]
```

### Function that extracts the province name from the "Lokalizacja" column

```python
def wyciagnij_wojewodztwo(tekst):
    for wojewodztwo in wojewodztwa:
        if wojewodztwo.lower() in tekst.lower():
            return wojewodztwo.capitalize()
    return "Nie znaleziono"
```

### Function that extracts the city name from the "Lokalizacja" column

```python
def wyciagnij_miasto(tekst):
    for miasto in miasta:
        if miasto.lower() in tekst.lower():
            return miasto.capitalize()
    return "Nie znaleziono"
```

### Function, which, based on the city from the city_village frame, defines the province

```python
def wojewodztwo_na_podstawie_miasta(tekst):
    for miasto in miasta:
        if miasto.lower() in tekst.lower(): 
            if len(miasta_wojewodztwa[miasta_wojewodztwa["nazwa miasta"] == miasto.capitalize()]) != 0:
                index_woj = miasta_wojewodztwa[miasta_wojewodztwa["nazwa miasta"] == miasto.capitalize()].index[0] - 1
                return miasta_wojewodztwa["nazwa województwa"].iloc[index_woj]
            else:
                return "Nie znaleziono"
        
    
```

### Function that returns text in which only the first letter is uppercase

```python
def cap(tekst):
    if pd.isna(tekst):
        return np.nan       
    else:
        return tekst.capitalize()
        
    
```

```python
def extract_date(date_str):
    date_obj = datetime.strptime(date_str, "%d/%m/%Y")
    return date_obj
```

```python
def extract_date_with_hour(date_str):
    # Parsowanie tekstu na obiekt daty
    date_obj = datetime.strptime(date_str, "%Y-%m-%d %H:%M:%S")
    return date_obj.date()
```

```python
def return_month(date_obj):
    return date_obj.month
```

```python
def return_year(date_obj):
    return date_obj.year
```

```python
def return_day(date_obj):
    return date_obj.day
```

```python
def spliting_features(x):
    y = x.split(",")
    y_list = []
    for element in y: 
        y_list.append(element.replace("[", "").replace("'", "").replace("]", "").replace(" ", ""))
    
    return y_list
```

```python
df3["Features"] = df3["Features"].apply(spliting_features)
```

```python
unique_list = []


def uni(x):
    for el in x:
        if el not in unique_list:
            unique_list.append(el)
```

```python
df3["Features"].apply(uni)
```

```python
unique_list = unique_list[1:]
```

```python
zero_data = np.zeros((208304, 70))
```

```python
uni_df = pd.DataFrame(columns=unique_list, data=zero_data)
```

```python
for i in range(0,len(df3["Features"])):
    for j in range(0, len(unique_list)):
        for k in range(0, len(df3["Features"].iloc[i])):
            if df3["Features"].iloc[i][k] == unique_list[j]:
                uni_df[df3["Features"].iloc[i][k]][i] = 1
    
    
        
                
                
    
    
```
