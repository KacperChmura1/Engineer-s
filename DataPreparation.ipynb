## Uzupełnianie spalania w mieście

```python
# Combustion in the city do based on the combustion of other cars with a similar engine and power
```

```python
df2["Pojemność skokowa"].value_counts()
```

```python
df2["Spalanie W Mieście"].value_counts()
```

```python
df2.columns
```

```python
#Imputacja danych na podstawie Pojemności skokowej
```

```python
import pandas as pd

# Grupowanie po wartościach w kolumnie "Pojemność skokowa" i obliczenie średniego "Spalania w mieście"
srednie_spalanie = df2.groupby('Pojemność skokowa')['Spalanie W Mieście'].mean()

# Uzyskanie listy indeksów brakujących wartości w kolumnie "Spalanie w mieście"
indeksy_brakujace = df2[df2['Spalanie W Mieście'].isnull()].index

# Iteracja przez indeksy brakujące i uzupełnienie ich na podstawie obliczonych średnich
for indeks in indeksy_brakujace:
    pojemnosc_skokowa = df2.at[indeks, 'Pojemność skokowa']
    if pojemnosc_skokowa in srednie_spalanie:
        srednia = srednie_spalanie[pojemnosc_skokowa]
        df2.at[indeks, 'Spalanie W Mieście'] = srednia
```

```python
plt.figure(figsize=(9, 3), dpi=150)
sns.heatmap(df2.isnull())
```

## Uzupełnianie liczba miejsc

```python
df2["Liczba miejsc"]
```

```python
df2[pd.isna(df2["Liczba miejsc"])]
```

```python
# Grupowanie po modelu pojazdu i obliczenie średnich liczby miejsc i liczby drzwi
srednie_wartosci = df2.groupby('Model pojazdu')[['Liczba miejsc', 'Liczba drzwi']].mean()

# Iteracja przez wiersze z brakującymi wartościami
for index, row in df2.iterrows():
    model_pojazdu = row['Model pojazdu']
    if pd.isnull(row['Liczba miejsc']):
        if model_pojazdu in srednie_wartosci.index:
            df2.at[index, 'Liczba miejsc'] = srednie_wartosci.loc[model_pojazdu]['Liczba miejsc']
    if pd.isnull(row['Liczba drzwi']):
        if model_pojazdu in srednie_wartosci.index:
            df2.at[index, 'Liczba drzwi'] = srednie_wartosci.loc[model_pojazdu]['Liczba drzwi']

```

```python
plt.figure(figsize=(9, 3), dpi=150)
sns.heatmap(df2.isnull())
```

## Wypełnienie tempomat

```python
df2[pd.isna(df2["Tempomat"])]["Rok produkcji"].value_counts()
```

```python
# Grupowanie po modelu pojazdu i roku produkcji oraz obliczenie średnich wartości tempomatu
srednie_tempomat = df2.groupby(['Model pojazdu', 'Rok produkcji'])['Tempomat'].mean()

# Iteracja przez wiersze z brakującymi wartościami
for index, row in df2.iterrows():
    model_pojazdu = row['Model pojazdu']
    rok_produkcji = row['Rok produkcji']
    if pd.isnull(row['Tempomat']):
        if (model_pojazdu, rok_produkcji) in srednie_tempomat.index:
            df2.at[index, 'Tempomat'] = srednie_tempomat.loc[(model_pojazdu, rok_produkcji)]
```
