## Uzupenianie spalania w miecie

```python
# Combustion in the city do based on the combustion of other cars with a similar engine and power
```

```python
df2["Pojemno skokowa"].value_counts()
```

```python
df2["Spalanie W Miecie"].value_counts()
```

```python
df2.columns
```

```python
#Imputacja danych na podstawie Pojemnoci skokowej
```

```python
import pandas as pd

# Grupowanie po wartociach w kolumnie "Pojemno skokowa" i obliczenie redniego "Spalania w miecie"
srednie_spalanie = df2.groupby('Pojemno skokowa')['Spalanie W Miecie'].mean()

# Uzyskanie listy indeksw brakujcych wartoci w kolumnie "Spalanie w miecie"
indeksy_brakujace = df2[df2['Spalanie W Miecie'].isnull()].index

# Iteracja przez indeksy brakujce i uzupenienie ich na podstawie obliczonych rednich
for indeks in indeksy_brakujace:
    pojemnosc_skokowa = df2.at[indeks, 'Pojemno skokowa']
    if pojemnosc_skokowa in srednie_spalanie:
        srednia = srednie_spalanie[pojemnosc_skokowa]
        df2.at[indeks, 'Spalanie W Miecie'] = srednia
```

```python
plt.figure(figsize=(9, 3), dpi=150)
sns.heatmap(df2.isnull())
```

## Uzupenianie liczba miejsc

```python
df2["Liczba miejsc"]
```

```python
df2[pd.isna(df2["Liczba miejsc"])]
```

```python
# Grupowanie po modelu pojazdu i obliczenie rednich liczby miejsc i liczby drzwi
srednie_wartosci = df2.groupby('Model pojazdu')[['Liczba miejsc', 'Liczba drzwi']].mean()

# Iteracja przez wiersze z brakujcymi wartociami
for index, row in df2.iterrows():
    model_pojazdu = row['Model pojazdu']
    if pd.isnull(row['Liczba miejsc']):
        if model_pojazdu in srednie_wartosci.index:
            df2.at[index, 'Liczba miejsc'] = srednie_wartosci.loc[model_pojazdu]['Liczba miejsc']
    if pd.isnull(row['Liczba drzwi']):
        if model_pojazdu in srednie_wartosci.index:
            df2.at[index, 'Liczba drzwi'] = srednie_wartosci.loc[model_pojazdu]['Liczba drzwi']

```

```python
plt.figure(figsize=(9, 3), dpi=150)
sns.heatmap(df2.isnull())
```

## Wypenienie tempomat

```python
df2[pd.isna(df2["Tempomat"])]["Rok produkcji"].value_counts()
```

```python
# Grupowanie po modelu pojazdu i roku produkcji oraz obliczenie rednich wartoci tempomatu
srednie_tempomat = df2.groupby(['Model pojazdu', 'Rok produkcji'])['Tempomat'].mean()

# Iteracja przez wiersze z brakujcymi wartociami
for index, row in df2.iterrows():
    model_pojazdu = row['Model pojazdu']
    rok_produkcji = row['Rok produkcji']
    if pd.isnull(row['Tempomat']):
        if (model_pojazdu, rok_produkcji) in srednie_tempomat.index:
            df2.at[index, 'Tempomat'] = srednie_tempomat.loc[(model_pojazdu, rok_produkcji)]
```
